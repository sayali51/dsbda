```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset from a direct URL for accessibility
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display the first 5 rows
print(df.head())
```

```python
# Check dimensions (Rows, Columns)
print(f"Dataset Shape: {df.shape}")

# Initial statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Summarize data types
print(df.dtypes)

# Example Conversion: Ensure 'Survived' is treated as a category/factor 
# and 'Fare' is a float (it usually is, but let's be sure).
df['Survived'] = df['Survived'].astype('category')
df['Fare'] = df['Fare'].astype('float')

# Filling missing values (Simple Normalization)
# We fill missing Age values with the median to avoid skewing the data
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Convert 'Sex' (male, female) into dummy/indicator variables
# This creates new columns: Sex_female and Sex_male
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)

# Explanation: 'drop_first=True' helps avoid the 'Dummy Variable Trap' 
# by representing Gender in one column (1 for male, 0 for female).

print(df.head())
```
