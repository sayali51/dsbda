```python
import pandas as pd
import numpy as np
```

```python
# Example data
data = {
    'AgeGroup': ['Young', 'Young', 'Middle', 'Middle', 'Old', 'Old', 'Young', 'Middle', 'Old'],
    'Income': [25000, 27000, 40000, 42000, 55000, 60000, 26000, 41000, 58000]
}

df = pd.DataFrame(data)
print(df)
```

```python
grouped = df.groupby('AgeGroup')['Income'].agg(['mean', 'median', 'min', 'max', 'std'])
print(grouped)
```

```python
# Dictionary: AgeGroup → list of incomes
income_list = df.groupby('AgeGroup')['Income'].apply(list).to_dict()
print(income_list)
```

```python
# Load iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
iris = pd.read_csv(url)
print(iris.head())
```

```python
setosa = iris[iris['species'] == 'setosa']
versicolor = iris[iris['species'] == 'versicolor']
virginica = iris[iris['species'] == 'virginica']
```

```python
print("Setosa Stats:")
print(setosa.describe())

print("\nVersicolor Stats:")
print(versicolor.describe())

print("\nVirginica Stats:")
print(virginica.describe())
```
